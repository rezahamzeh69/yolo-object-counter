```python
# @title YOLO Object Counter (v8+) - Colab (Integrated)

try:
    import ultralytics
    print(f"Ultralytics {ultralytics.__version__} already installed.")
except ImportError:
    print("Installing Ultralytics...")
    !pip install ultralytics -q
    import ultralytics
    print(f"Ultralytics {ultralytics.__version__} installed.")

import cv2
from ultralytics import YOLO
import collections
import time
import os
import torch
import traceback
from google.colab import files
from google.colab.patches import cv2_imshow
from IPython.display import display, Javascript, Image
import numpy as np
from base64 import b64decode
import pkg_resources

def get_yolo_versions():
    yolo_versions = []
    try:
        yolo_module = __import__('ultralytics.models')
        models_dir = os.path.dirname(yolo_module.models.__file__)
        potential_dirs = [d for d in os.listdir(models_dir) if os.path.isdir(os.path.join(models_dir, d)) and (d.startswith('yolo') or (d.startswith('v') and len(d) > 1 and d[1].isdigit()))]
        for d in potential_dirs:
             if d.startswith('v') and d[1:].isdigit() and int(d[1:]) >= 8:
                 yolo_versions.append(f"v{d[1:]}")
        known_versions = ['v8']
        all_found = sorted(list(set(known_versions + yolo_versions)), key=lambda x: int(x[1:]))
        if not all_found: return ['v8']
        return all_found
    except Exception:
        print("Warning: Could not dynamically detect YOLO versions, using default [v8].")
        return ['v8']

DETECTED_YOLO_VERSIONS = get_yolo_versions()
LATEST_YOLO_VERSION_NUM = max([int(v[1:]) for v in DETECTED_YOLO_VERSIONS])

AVAILABLE_MODELS = {}
for ver_str in DETECTED_YOLO_VERSIONS:
    ver_num = int(ver_str[1:])
    suffixes = ['n', 's', 'm', 'l', 'x']
    AVAILABLE_MODELS[ver_str] = [f"yolo{ver_str}{suffix}" for suffix in suffixes]

DEFAULT_MODEL_BASE = f'yolov{LATEST_YOLO_VERSION_NUM}m'

# --- Settings ---
print("\n--- Settings ---")
SELECTED_YOLO_VERSION = f'v{LATEST_YOLO_VERSION_NUM}'
MODEL_NAME = DEFAULT_MODEL_BASE #@param ['yolov8n', 'yolov8s', 'yolov8m', 'yolov8l', 'yolov8x'] # Default overridden by logic
CONFIDENCE_THRESHOLD = 0.30 #@param {type:"slider", min:0.1, max:0.9, step:0.05}
TARGET_CLASSES = None
INPUT_SIZE = 640 #@param [640, 1280] {type:"raw"}
SKIP_FRAMES_FACTOR = 1 #@param {type:"integer"}
USE_TTA = False #@param {type:"boolean"}

# --- Global Vars ---
model = None
class_names = {}
DEVICE = 'cpu'
current_model_name = ""

def check_cuda():
    global DEVICE
    detected_device = 'cpu'
    try:
        if torch.cuda.is_available():
            gpu_name = torch.cuda.get_device_name(0)
            print(f"‚úÖ CUDA (GPU) available: {gpu_name}")
            detected_device = 'cuda'
        else:
            print("‚ö†Ô∏è CUDA not available, using CPU.")
    except Exception as e:
        print(f"Error checking CUDA: {e}")
    finally:
        DEVICE = detected_device
        print(f"Processing device: '{DEVICE}'")

def load_yolo_model(model_full_name_to_load):
    global model, class_names, DEVICE, current_model_name
    model_pt_filename = model_full_name_to_load + ".pt"

    if model is not None and current_model_name == model_full_name_to_load:
        print(f"‚úÖ Model '{current_model_name}' is already loaded.")
        return True

    if model is not None:
        print(f"Releasing previous model ({current_model_name})...")
        try:
            del model
            if DEVICE == 'cuda': torch.cuda.empty_cache()
            model = None; class_names = {}; current_model_name = ""
        except Exception as e_del: print(f"Error releasing model: {e_del}")

    try:
        print(f"‚è≥ Loading/Downloading model '{model_pt_filename}'...")
        model_instance = YOLO(model_pt_filename)
        model_instance.to(DEVICE)
        print(f"   Model transferred to '{DEVICE}'.")
        print("   Warming up model...")
        dummy_input = torch.zeros(1, 3, INPUT_SIZE, INPUT_SIZE).to(DEVICE)
        _ = model_instance.predict(dummy_input, verbose=False, device=DEVICE)
        model = model_instance
        class_names = model.names
        current_model_name = model_full_name_to_load
        print(f"‚úÖ Model '{current_model_name}' loaded successfully ({len(class_names)} classes).")
        return True
    except Exception as e:
        print(f"‚ùå Error loading/downloading model '{model_pt_filename}': {e}")
        model = None; class_names = {}; current_model_name = ""
        return False

def process_predictions(results):
    total_count = 0
    specific_counts = collections.defaultdict(int)
    if not class_names: return 0, specific_counts
    if results.boxes is not None:
        for box in results.boxes:
            if box.cls is not None and box.conf is not None and len(box.cls) > 0 and len(box.conf) > 0:
                try:
                    cls_id = int(box.cls[0].item())
                    confidence = float(box.conf[0].item())
                    if confidence >= CONFIDENCE_THRESHOLD:
                        class_name = class_names.get(cls_id, f"ID:{cls_id}")
                        if TARGET_CLASSES is None or class_name in TARGET_CLASSES:
                            specific_counts[class_name] += 1
                            total_count += 1
                except: pass
    return total_count, specific_counts

def annotate_frame(frame, total_count, specific_counts, fps=None):
    y_offset = 30; font_scale_large = 0.7; font_scale_small = 0.6
    thickness = 2; font = cv2.FONT_HERSHEY_SIMPLEX
    color_total=(0,0,255); color_specific=(0,255,0); color_fps=(255,0,0)
    bg_color=(0,0,0); text_margin=5
    target_display = 'All' if TARGET_CLASSES is None else ','.join(TARGET_CLASSES)
    text_total = f"Total ({target_display}): {total_count}"
    (w_total, h_total), _ = cv2.getTextSize(text_total, font, font_scale_large, thickness)
    cv2.rectangle(frame, (5, y_offset - h_total - text_margin), (5 + w_total + text_margin, y_offset + text_margin), bg_color, -1)
    cv2.putText(frame, text_total, (10, y_offset), font, font_scale_large, color_total, thickness, cv2.LINE_AA)
    y_offset += h_total + 15
    sorted_counts = sorted(specific_counts.items())
    for name, count in sorted_counts:
        text_specific = f"{name}: {count}"
        (w_spec, h_spec), _ = cv2.getTextSize(text_specific, font, font_scale_small, thickness)
        cv2.rectangle(frame, (5, y_offset - h_spec - text_margin), (5 + w_spec + text_margin, y_offset + text_margin), bg_color, -1)
        cv2.putText(frame, text_specific, (10, y_offset), font, font_scale_small, color_specific, thickness, cv2.LINE_AA)
        y_offset += h_spec + 10
        if y_offset > frame.shape[0] - 30: break
    if fps is not None:
        text_fps = f"FPS: {fps:.1f} ({DEVICE.upper()})"
        (w_fps, h_fps), _ = cv2.getTextSize(text_fps, font, font_scale_small, thickness)
        fps_x = frame.shape[1] - w_fps - 10
        cv2.rectangle(frame, (fps_x - text_margin, 5), (frame.shape[1] - 5, 5 + h_fps + text_margin), bg_color, -1)
        cv2.putText(frame, text_fps, (fps_x, 5 + h_fps), font, font_scale_small, color_fps, thickness, cv2.LINE_AA)
    return frame

def process_uploaded_image():
    global USE_TTA
    if not model: print("‚ùå Model not loaded."); return
    print("\n--- Process Image ---")
    print("üìÇ Please upload your image file...")
    uploaded = files.upload()
    if not uploaded: print("No file uploaded."); return
    filename = list(uploaded.keys())[0]
    print(f"\nProcessing image: {filename}")
    try:
        img_bytes = uploaded[filename]
        nparr = np.frombuffer(img_bytes, np.uint8)
        img = cv2.imdecode(nparr, cv2.IMREAD_COLOR)
        if img is None: print(f"‚ùå Error reading image: {filename}"); return
        print(f"   Using TTA: {USE_TTA}")
        start_time = time.time()
        results_list = model.predict(img, stream=False, conf=CONFIDENCE_THRESHOLD, imgsz=INPUT_SIZE, device=DEVICE, verbose=False, augment=USE_TTA)
        end_time = time.time()
        if results_list:
            results = results_list[0]
            annotated_img_boxes = results.plot(conf=True, line_width=2, font_size=None, labels=True)
            total_count, specific_counts = process_predictions(results)
            final_annotated_img = annotate_frame(annotated_img_boxes, total_count, specific_counts)
            processing_time = end_time - start_time
            print(f"\nüìä Results ({processing_time:.2f} sec):")
            print(f"   Total Count: {total_count}")
            if specific_counts:
                for name, count in sorted(specific_counts.items()): print(f"     - {name}: {count}")
            else: print("   No target objects detected.")
            print("\nüñºÔ∏è Displaying processed image:")
            cv2_imshow(final_annotated_img)
        else: print("Processing yielded no results.")
    except Exception as e: print(f"‚ùå Error processing image: {e}"); traceback.print_exc(limit=1)

def process_uploaded_video():
    global USE_TTA
    if not model: print("‚ùå Model not loaded."); return
    print("\n--- Process Video ---")
    print("üìÇ Please upload your video file (can take time)...")
    uploaded = files.upload()
    if not uploaded: print("No file uploaded."); return
    input_filename = list(uploaded.keys())[0]
    output_filename = f"processed_{os.path.splitext(input_filename)[0]}.mp4"
    output_path = os.path.join("/content/", output_filename)
    print(f"\nProcessing video: {input_filename}")
    print(f"Output: {output_path}")
    cap = None; writer = None; frame_count = 0; total_frames_processed = 0
    start_total_time = time.time()
    temp_input_path = os.path.join("/content/", input_filename)
    try:
        with open(temp_input_path, 'wb') as f: f.write(uploaded[input_filename])
        cap = cv2.VideoCapture(temp_input_path)
        if not cap.isOpened(): print(f"‚ùå Error opening video: {input_filename}"); return
        fw = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH)); fh = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
        fps = cap.get(cv2.CAP_PROP_FPS); total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
        print(f"   Info: {fw}x{fh} @ {fps:.2f} FPS (~{total_frames} frames)")
        print(f"   Using TTA: {USE_TTA}")
        fourcc = cv2.VideoWriter_fourcc(*'mp4v')
        writer = cv2.VideoWriter(output_path, fourcc, fps, (fw, fh))
        if not writer.isOpened(): raise IOError("Could not open VideoWriter")
        print("\n‚è≥ Starting frame processing...")
        prev_time = time.time(); accumulated_time_fps = 0; frame_count_fps = 0; display_fps = 0
        while True:
            ret, frame = cap.read()
            if not ret: print("\nüèÅ End of video."); break
            frame_count += 1
            if frame_count % SKIP_FRAMES_FACTOR != 1 and SKIP_FRAMES_FACTOR > 1:
                writer.write(frame); continue
            total_frames_processed += 1
            current_time = time.time(); elapsed = current_time - prev_time; prev_time = current_time
            accumulated_time_fps += elapsed; frame_count_fps += 1
            if accumulated_time_fps >= 1.0: display_fps = frame_count_fps / accumulated_time_fps; accumulated_time_fps = 0; frame_count_fps = 0
            results_list = model.predict(frame, stream=False, conf=CONFIDENCE_THRESHOLD, imgsz=INPUT_SIZE, device=DEVICE, verbose=False, augment=USE_TTA)
            annotated_frame = frame; total_count = 0; specific_counts = collections.defaultdict(int)
            if results_list:
                results = results_list[0]
                annotated_frame_boxes = results.plot(conf=True, line_width=1, labels=True)
                total_count, specific_counts = process_predictions(results)
                annotated_frame = annotate_frame(annotated_frame_boxes, total_count, specific_counts, display_fps)
            writer.write(annotated_frame)
            if frame_count % 50 == 0: print(f"   Frame {frame_count}... (FPS:{display_fps:.1f})")
        end_total_time = time.time(); total_processing_time = end_total_time - start_total_time
        print(f"\n‚úÖ Video processing complete ({total_processing_time:.2f} sec).")
        print(f"   Frames processed (skipped accounted): {total_frames_processed}/{frame_count}")
        print(f"   Output file: {output_path}")
        print("\nüîó Creating download link...")
        files.download(output_path)
        print("   Download link should appear.")
    except Exception as e: print(f"‚ùå Error processing video: {e}"); traceback.print_exc(limit=1)
    finally:
        if cap: cap.release()
        if writer: writer.release()
        if os.path.exists(temp_input_path):
             try: os.remove(temp_input_path)
             except OSError as e: print(f"   Error removing temp file: {e}")
        print("   Resources released.")

def take_photo(filename='photo_webcam.jpg', quality=0.8):
  js = Javascript('''
    async function takePhoto(quality) {
      const div = document.createElement('div');
      const capture = document.createElement('button');
      capture.textContent = 'Capture';
      div.appendChild(capture);
      const video = document.createElement('video');
      video.style.display = 'block';
      const stream = await navigator.mediaDevices.getUserMedia({video: true});
      const outputContainer = document.getElementById('output-area');
      if (outputContainer) { outputContainer.appendChild(div); }
      else { document.body.appendChild(div); }
      div.appendChild(video);
      video.srcObject = stream;
      await video.play();
      google.colab.output.setIframeHeight(document.documentElement.scrollHeight, true);
      await new Promise((resolve) => capture.onclick = resolve);
      const canvas = document.createElement('canvas');
      canvas.width = video.videoWidth; canvas.height = video.videoHeight;
      canvas.getContext('2d').drawImage(video, 0, 0);
      stream.getVideoTracks()[0].stop();
      div.remove();
      return canvas.toDataURL('image/jpeg', quality);
    }
    ''')
  display(js)
  try:
      from google.colab.output import eval_js
      data = eval_js('takePhoto({})'.format(quality))
      binary = b64decode(data.split(',')[1])
      with open(filename, 'wb') as f: f.write(binary)
      return filename
  except Exception as e:
      print(f"Error communicating with webcam JavaScript: {e}")
      return None

def process_webcam_capture():
    global USE_TTA
    if not model: print("‚ùå Model not loaded."); return
    print("\n--- Process Webcam Frame ---")
    try:
        print("üì∑ Preparing webcam... Please grant permission and click 'Capture'.")
        filename = take_photo()
        if filename is None: return
        print(f"\n‚úÖ Image saved to '{filename}'.")
        img = cv2.imread(filename)
        if img is None: print("‚ùå Error reading captured image."); return
        print("   Processing webcam image...")
        print(f"   Using TTA: {USE_TTA}")
        start_time = time.time()
        results_list = model.predict(img, stream=False, conf=CONFIDENCE_THRESHOLD, imgsz=INPUT_SIZE, device=DEVICE, verbose=False, augment=USE_TTA)
        end_time = time.time()
        if results_list:
            results = results_list[0]
            annotated_img_boxes = results.plot(conf=True, line_width=2, labels=True)
            total_count, specific_counts = process_predictions(results)
            final_annotated_img = annotate_frame(annotated_img_boxes, total_count, specific_counts)
            processing_time = end_time - start_time
            print(f"\nüìä Results ({processing_time:.2f} sec):")
            print(f"   Total Count: {total_count}")
            if specific_counts:
                for name, count in sorted(specific_counts.items()): print(f"     - {name}: {count}")
            else: print("   No target objects detected.")
            print("\nüñºÔ∏è Displaying processed webcam image:")
            cv2_imshow(final_annotated_img)
        else: print("Processing yielded no results.")
        if os.path.exists(filename):
             try: os.remove(filename)
             except OSError: pass
    except ImportError: print("‚ùå Import Error: Not running in Colab?")
    except Exception as e: print(f"‚ùå Error processing webcam: {e}"); traceback.print_exc(limit=1)

# --- Main Execution ---
check_cuda()
print(f"\nInitial model selection: {MODEL_NAME}")
print(f"Confidence threshold: {CONFIDENCE_THRESHOLD}")
print(f"Input size: {INPUT_SIZE}")
print(f"Using TTA: {USE_TTA}")

initial_model_to_load = MODEL_NAME
if load_yolo_model(initial_model_to_load):
    while True:
        print("\n" + "="*40)
        print(f"   Current Model: {current_model_name} ({DEVICE.upper()}) | Conf: {CONFIDENCE_THRESHOLD} | Size: {INPUT_SIZE} | TTA: {USE_TTA}")
        print("   Select operation:")
        print("     1. Process New Image (Upload)")
        print("     2. Process New Video (Upload)")
        print("     3. Process Webcam Frame")
        print("     4. Change YOLO Model")
        print("     5. Change Settings (Conf, Size, TTA)")
        print("     0. Exit")
        print("="*40)
        choice = input("Enter operation number (0-5): ")

        if choice == '1':
            process_uploaded_image()
        elif choice == '2':
            process_uploaded_video()
        elif choice == '3':
            process_webcam_capture()
        elif choice == '4':
            print("\n--- Change Model ---")
            all_models_flat = [item for sublist in AVAILABLE_MODELS.values() for item in sublist]
            print("Available models:")
            for i, m_name in enumerate(all_models_flat): print(f"  {i+1}. {m_name}")
            try:
                model_choice_idx = int(input(f"Enter model number (1-{len(all_models_flat)}): ")) - 1
                if 0 <= model_choice_idx < len(all_models_flat):
                    selected_model_full_name = all_models_flat[model_choice_idx]
                    print(f"\nAttempting to load new model: {selected_model_full_name}")
                    load_yolo_model(selected_model_full_name)
                else:
                    print("‚ùå Invalid number.")
            except ValueError:
                print("‚ùå Invalid input. Please enter a number.")
        elif choice == '5':
             print("\n--- Change Settings ---")
             try:
                 new_conf_str = input(f"New confidence threshold (current: {CONFIDENCE_THRESHOLD}): ").strip()
                 if new_conf_str: # Only update if user entered something
                     new_conf = float(new_conf_str)
                     if 0.05 <= new_conf <= 0.95: CONFIDENCE_THRESHOLD = new_conf
                     else: print("Confidence must be between 0.05 and 0.95.")
             except ValueError: print("Invalid confidence input.")
             try:
                 new_size_str = input(f"New input size (current: {INPUT_SIZE}, options: 640, 1280): ").strip()
                 if new_size_str:
                     new_size = int(new_size_str)
                     if new_size in [640, 1280]: INPUT_SIZE = new_size
                     else: print("Input size must be 640 or 1280.")
             except ValueError: print("Invalid size input.")
             try:
                 tta_choice = input(f"Use TTA (Test-Time Augmentation)? (current: {USE_TTA}) [y/n]: ").strip().lower()
                 if tta_choice == 'y': USE_TTA = True
                 elif tta_choice == 'n': USE_TTA = False
                 elif tta_choice: print("Invalid TTA input (use y or n).") # Only print error if input given
             except ValueError: print("Invalid TTA input.")
             print("Settings updated.")
        elif choice == '0':
            print("\nExiting program...")
            break
        else:
            print("‚ùå Invalid choice. Please enter a number between 0 and 5.")

        if choice != '0':
            try:
                input("\n[Press Enter to continue...]")
            except EOFError:
                 print("\n[Auto-continuing...]")
                 time.sleep(1)
else:
    print("\n‚ùå Initial model loading failed. Cannot continue.")

print("\n--- Cell execution finished ---")
```
